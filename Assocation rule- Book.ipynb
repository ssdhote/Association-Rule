### lift<1 suggest that antecedents has decreased probability that of consequents by factor numerically equal to lift
### lift=1 suggest that antecedents and consequents independent of each other

## Removing Duplicate Rules

```python
def to_list(i):
    return (sorted(list(i)))
```

```python
X = rules.antecedents.apply(to_list)+rules.consequents.apply(to_list)


X = X.apply(sorted)

rules2_sets = list(X)

unique_rules_sets = [list(m) for m in set(tuple(i) for i in rules2_sets)]
index_rules2 = []
for i in unique_rules_sets:
    index_rules2.append(rules2_sets.index(i))
    
rules2_no_redudancy  = rules.iloc[index_rules2,:]
rules2_no_redudancy.shape
#396 rules
rules2_no_redudancy.sort_values('lift',ascending=False).head(10)
```
