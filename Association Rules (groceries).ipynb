```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
groceries= pd.read_csv("C:\\Users\\DELL\\Downloads\\groceries.csv",error_bad_lines=False)
groceries.head()
```

```python
groceries.isnull().sum()
```

```python
groceries["margarine"]=groceries.fillna(groceries.mode().iloc[:,1:2])
groceries.isnull().sum()
```

```python
groceries["ready soups"]=groceries.fillna(groceries.mode().iloc[:,2:3])
groceries.isnull().sum()
```

```python
groceries["semi-finished bread"]=groceries.fillna(groceries.mode().iloc[:,0:1])
groceries.isnull().sum()
```

```python
groceries_new=pd.get_dummies(groceries)
groceries_new.head()
```

```python
groceries_new.describe()
```

```python
frequent_itemsets = apriori(groceries_new, min_support=0.01, use_colnames=True)
frequent_itemsets.head()
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.75)
rules
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.75)
rules
```

```python
rules[rules.lift>1]
```

## Removing Duplicate Rules

```python
def to_list(i):
    return (sorted(list(i)))
```

```python
X = rules.antecedents.apply(to_list)+rules.consequents.apply(to_list)
X = X.apply(sorted)
rules_sets = list(X)
unique_rules_sets = [list(m) for m in set(tuple(i) for i in rules_sets)]
index_rules = []
for i in unique_rules_sets:
    index_rules.append(rules_sets.index(i))
    
rules_no_redudancy  = rules.iloc[index_rules,:]
rules_no_redudancy.shape
#396 rules
rules_no_redudancy.sort_values('lift',ascending=False).head(10)
```
